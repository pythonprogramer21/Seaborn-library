```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import norm
```

```python
a=sns.load_dataset("tips")
a
```

```python
sns.distplot(a['total_bill'])
plt.show()
```

```python
sns.distplot(a["tip"])
```

```python
sns.distplot(a["size"])
```

```python
sns.distplot(a['total_bill'],bins=78)
```

```python
sns.distplot(a['total_bill'],hist=False)
```

```python
sns.distplot(a['total_bill'],kde=False)
```

```python
sns.distplot(a['total_bill'],rug=True)
```

```python
sns.distplot(a['total_bill'],fit=norm)
```

```python
sns.distplot(a['total_bill'],fit=norm,kde=False)
```

```python
sns.distplot(a['total_bill'],fit=norm,color='r')
```

```python
sns.distplot(a['total_bill'],fit=norm,vertical=True)
plt.show()
```

```python
sns.distplot(a['total_bill'],fit=norm,norm_hist=True)
```

```python
sns.distplot(a['total_bill'],fit=norm,kde=False,axlabel="XYZ")
```

```python
sns.distplot(a['total_bill'],fit=norm,kde=False,label="Total bill")
plt.legend()
```

```python
a.total_bill.sort_values()
```

```python
sns.set()
plt.figure(figsize=(16,8))
bin=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(a["total_bill"],bins=bin)
plt.show()
```

```python
sns.set()
plt.figure(figsize=(16,8))
bin=[1,5,10,15,20,25,30,35,40,45,50,55]
sns.distplot(a["total_bill"],bins=bin)
plt.xticks(bin)   
plt.show()
```

```python

plt.figure(figsize=(16,8))
sns.distplot(a["total_bill"],hist_kws={'color':'r','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':.9})
```

```python
plt.figure(figsize=(16,8))
sns.distplot(a["total_bill"],hist_kws={'color':'r','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':.9},kde_kws={'color':'r','linewidth':5,'linestyle':'--','alpha':.9})
```

```python
plt.figure(figsize=(16,8))
sns.distplot(a["total_bill"],hist_kws={'color':'r','edgecolor':'y','linewidth':5,'linestyle':'--','alpha':.9},kde_kws={'color':'r','linewidth':5,'linestyle':'--','alpha':.9},rug=True,rug_kws={'color':'r','linewidth':5,'linestyle':'--','alpha':.9})
```

```python
sns.distplot(a['total_bill'],kde=False,fit=norm,fit_kws={'color':'r','linewidth':5,'linestyle':'--','alpha':.9})
plt.show()
```
