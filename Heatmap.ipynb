```python
r=pd.read_csv("F:\\suryakant notes\\Who_is_responsible_for_global_warming.csv")
r
```

```python
d=r.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index('Country Name')
d.head(7)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,cmap='coolwarm')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,cmap='coolwarm',annot=True)
```

```python
plt.figure(figsize=(15,8))
aal={'fontsize':10,'fontstyle':'italic','color':'k','alpha':0.6,'verticalalignment':'center','backgroundcolor':'w'}
sns.heatmap(d,cmap='coolwarm',annot=True,annot_kws=aal)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,cmap='coolwarm',annot=True,linewidth=4,linecolor='k')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,cbar=False)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(d,cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(15,8))
kl={'orientation':'horizontal','shrink':1,'extend':'min','extendfrac':0.1,'ticks':np.arange(0,22),'drawedges':True}
sns.heatmap(d,cbar_kws=kl)
```

```python
plt.figure(figsize=(15,8))
ax=sns.heatmap(d)
ax.set(title='Who_is_responsible_for_global_warming',xlabel='Year',ylabel='Name')
```

```python
plt.figure(figsize=(15,8))
ax=sns.heatmap(d)
ax.set(title='Who_is_responsible_for_global_warming',xlabel='Year',ylabel='Country Name')
sns.set(font_scale=4)
```
