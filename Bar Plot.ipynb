```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a=sns.load_dataset("tips")
a
```

```python
sns.barplot(x=a.day,y=a.total_bill)
```

```python
sns.barplot(x=a['day'],y=a['total_bill'])
```

```python
sns.barplot(x=a.day,y=a.total_bill,hue=a.sex)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=a)
```

```python
kq=['Sun','Thur','Fri','Sat']
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=a,order=kq)
```

```python
dk=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=a,hue_order=dk)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='day',y='total_bill',hue='sex',data=a,estimator=np.max)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='day',y='total_bill',hue='sex',data=a,estimator=np.max,ci=56)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='day',y='total_bill',hue='sex',data=a,estimator=np.max,n_boot=2)
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='day',y='total_bill',hue='sex',data=a,estimator=np.max,orient='v')
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='tip',y='total_bill',data=a,hue='sex',estimator=np.max,orient='h')
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=a,palette='hot')
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=a,palette='magma')
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=a,saturation=3)
```
