```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a=sns.load_dataset("titanic")
a
```

```python
sns.scatterplot(x='age',y='fare',data=a)
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x='embark_town',y='fare',data=a,hue='sex')
```

```python
sns.scatterplot(x='age',y='sex',data=a)
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x='age',y='fare',data=a,hue='sex',style='who')
```

```python
plt.figure(figsize=(14,8))
sns.scatterplot(x='age',y='fare',data=a,hue='sex',style='who',size='who')
```
