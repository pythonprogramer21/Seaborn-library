```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
day=[12,67,34,35,22]
tem=[88,54,43,23,78]
d=pd.DataFrame({'day':day,'Tem':tem})
sns.lineplot(x='day',y='Tem',data=d)
plt.show()
```

```python
a=sns.load_dataset("tips")
a
```

```python
sns.lineplot(x='total_bill',y='tip',data=a)
```

```python
sns.lineplot(x='tip',y='total_bill',data=a)
```

```python
sns.lineplot(x='total_bill',y='size',data=a)
```

```python
sns.lineplot(x='size',y='total_bill',data=a)
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex')
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex')
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot')
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4)
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'])
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'],legend=False)
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'],legend='full')
```

```python
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'],legend='brief')
plt.title('Line Plot',fontsize=20)
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total bill',fontsize=15)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'],legend='brief')
plt.title('Line Plot',fontsize=20)
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total bill',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=a,hue='sex',style='sex',palette='hot',dashes=False,size=4,markers=['o','>'],legend='brief')
plt.title('Line Plot',fontsize=20)
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total bill',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=a,hue='day',style='day',palette='hot',dashes=False,size=4,markers=['o','>','<','^'],legend='brief')
plt.title('Line Plot',fontsize=20)
plt.xlabel('Size',fontsize=15)
plt.ylabel('Total bill',fontsize=15)
plt.show()
```
